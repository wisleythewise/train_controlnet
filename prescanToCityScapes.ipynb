```python
# Get an image from a file and change the colors of the image 
from PIL import Image

# Open the image file


# Define a function to map the colors of the image
def map_rgb(r, g, b):
    t = 80 

    neural = (68,1,84)
    
    # Sky
    is_sky = 128 - t < r < 128 + t and \
            4 - t < g < 4 + t and \
            252 - t < b < 252+ t 
    if is_sky:
        return  neural

    # Car
    is_car = 255 - t < r < 255 + t and \
            255 - t < g < 255 + t and \
            255 - t < b < 255+ t 
    
    if is_car:
        return  (71, 192, 110)

    # Truck
    is_truck = 1 - t < r < 1 + t and \
            255 - t < g < 255 + t and \
            1 - t < b < 1+ t 
    
    if is_truck:
        return  228, 227, 24


    # Road markings
    is_road_markings = 255 - t < r < 255 + t and \
            255 - t < g < 255 + t and \
            0 - t < b < 0+ t 
    
    if is_road_markings:
        return  253,231,36
    
    # Human
    is_human = 0 - t < r < 0 + t and \
            0 - t < g < 0 + t and \
            255 - t < b < 255+ t 
    
    if is_human:
        return  71, 30, 112

    # Road
    is_road = 126 - t < r < 126 + t and \
            126 - t < g < 126 + t and \
            126 - t < b < 126+ t 
    
    if is_road:
        return  253,231,36



    # Return the new colors
    return neural

# Get all teh images in segmentation folder
import os
path = "D:/nulimages_data/segmentation_crop/"
files = os.listdir(path)


for i, file in enumerate(files):
    im = Image.open(path + file)

    # Get the size of the image
    width, height = im.size

    # Process every pixel
    for x in range(width):
        for y in range(height):
            # Get the color of the pixel
            r, g, b = im.getpixel((x, y))

            # Change the color of the pixel
            r, g, b = map_rgb(r, g, b)
            # Save the changed pixel
            im.putpixel((x, y), (r, g, b))

    # Save the changed image
    im.save(f"D:/nulimages_data/segmentation_converted/{i}.png")
```
